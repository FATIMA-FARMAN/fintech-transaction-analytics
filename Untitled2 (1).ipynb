```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.pdfgen import canvas
```

```python
# Example dataset (you can upload your own)
data = {
    "TransactionID": [1,2,3,4,5,6],
    "Date": ["2025-08-01","2025-08-02","2025-08-02","2025-08-03","2025-08-04","2025-08-04"],
    "Amount": [120.5, 230.0, 75.0, 640.0, 310.0, 99.5],
    "Category": ["Food","Shopping","Transport","Electronics","Bills","Food"],
    "Payment_Status": ["Success","Failed","Success","Success","Pending","Success"],
    "Payment_Method": ["Card","Card","Cash","Card","Bank Transfer","Card"]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# Handle missing values
df = df.dropna()

# Normalize categories
df['Category'] = df['Category'].str.title()

# Sort by date
df = df.sort_values(by="Date")
df.head()
```

```python
# Total Spending
total_spent = df[df["Payment_Status"]=="Success"]["Amount"].sum()
print("Total Spent: $", total_spent)

# Spending by category
category_summary = df.groupby("Category")["Amount"].sum().reset_index()
category_summary
```

```python
# Spending by category
plt.figure(figsize=(8,5))
sns.barplot(data=category_summary, x="Category", y="Amount", palette="Blues")
plt.title("Spending by Category")
plt.show()

# Payment method breakdown
plt.figure(figsize=(6,6))
df["Payment_Method"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()
```

```python
from reportlab.lib.pagesizes import letter

c = canvas.Canvas("Transaction_Report.pdf", pagesize=letter)
c.drawString(100, 750, "📊 Fintech Transaction Analytics Report")
c.drawString(100, 730, f"Total Spending: ${total_spent}")
c.drawString(100, 710, "Top Spending Categories:")

y = 690
for i, row in category_summary.iterrows():
    c.drawString(120, y, f"{row['Category']}: ${row['Amount']}")
    y -= 20

c.save()
print("✅ PDF Report Generated: Transaction_Report.pdf")
```
